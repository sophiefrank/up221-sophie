# Open Spaces and Parks in the City of Los Angeles

This dataset contains information on the location of all open spaces and parks managed by the City of Los Angeles.

I start this data exploration by importing pandas.

```python
import pandas as pd
```

I then import the dataset containing information about open spaces and parks from Los Angeles GeoHub.

```python
park = pd.read_csv('Land_Use_and_Zoning.csv')
```

I look at the type of data using the type() command.

```python
type(park)
```

This is a data frame. To see how many columns and rows there are, I use the .shape command.

```python
park.shape
```

There are 522 rows and 55 columns in this data frame. To see more information about the data frame, I use the .info() command.

```python
park.info()
```

I can see there are 3 data types in this data frame: floats, integers, and objects. I then use the .head() command to look at the first 5 rows in the data frame.

```python
park.head()
```

There are quite a few columns in this data frame, so I trim the data so that I only see the columns I am interested in.

```python
new_columns = ['PARK_NAME','RPT_ACRES','CITY','TYPE','CENTER_LAT','CENTER_LON']
park[new_columns].sample(5)
```

Now only 6 columns are visible. Next I will declare this dataframe as a new variable.

```python
park_trimmed = park[new_columns].copy()
park_trimmed
```

I'd also like to rename some of these columns.

```python
park_trimmed.columns.to_list()
```

```python
park_trimmed.columns = ['Park Name', 'Acres', 'City', 'Type', 'Lat', 'Lon']
```

```python
park_trimmed
```

Now I'd like to see how many parks managed by the City of Los Angeles are in each city listed in this data frame. I use the .value_counts() command to do this.

```python
park_trimmed['City'].value_counts()
```

I only want to show parks in LA that are managed by the City of LA, so I will filter the data.

```python
park_trimmed[park_trimmed.City == 'Los Angeles']
```

I can see from the .value_count() command that there are 280 parks in Los Angeles and the data frame above has 280 rows. Now I will create a new data frame that only shows these parks.

```python
LA_park = park_trimmed[park_trimmed.City == 'Los Angeles'].copy()
```

```python
LA_park
```

Now I'm interested in seeing how many different types of parks there are in this data frame. I use the .value_counts() command for this.

```python
LA_park['Type'].value_counts()
```

Now I can see there are 4 types of parks in this data frame and most of them are categorized as Local Parks. I'd like to visualize this data, so I will use the .plot() command. 

```python
type_count = LA_park['Type'].value_counts()
```

```python
type_count.plot()
```

A bar chart would represent this data more clearly, so finally I use the .plot.bar() command to create this.

```python
type_count.plot.bar(x = 'Type', y = 'count', title = 'Number of Parks by Type')
```
